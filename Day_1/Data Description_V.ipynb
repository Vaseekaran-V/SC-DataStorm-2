### Getting info on categorical data

```python
def get_categorical_info(dataset):
    train_ = dataset.select_dtypes(exclude=['int', 'float'])
    for col in train_.columns:
        print(train_[col].unique()) # to print categories name only
        print(train_[col].value_counts()) # to print count of every category
        print()
```

```python
get_categorical_info(train_data)
```

```python
get_categorical_info(validation_data)
```

```python
get_categorical_info(test_data)
```
